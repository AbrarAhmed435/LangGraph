```python
from langgraph.graph import StateGraph ,START, END
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict
load_dotenv()
```

```python
model=ChatOpenAI(model='gpt-4o-mini')
```

```python
class BlogState(TypedDict):
    title:str
    outline:str
    content:str
    score:int
```

```python
graph=StateGraph(BlogState)
```

```python
def create_outline(state:BlogState)->BlogState:
    title=state['title']

    prompt=f"Genearte and outline for a blog on the topic : {title}"

    outline=model.invoke(prompt).content

    state['outline']=outline

    return state
```

```python
def create_blog(state:BlogState)->BlogState:
    title=state['title']
    outlline=state['outline']

    prompt=f"Genrate a detailed blog for topic: {title} with outline {outlline}"

    content=model.invoke(prompt).content
    state['content']=content
    return state
```

```python
from pydantic import BaseModel, Field
from typing import Annotated

class Score(BaseModel):
    score:Annotated[int,Field(description="Number between 1 and 10")]
```

```python
def generate_score(state:BlogState)->BlogState:
    content=state['content']
    prompt=f"Give a score of 1-10 for this blog  Blog: {content}"
    structured_model=model.with_structured_output(Score)
    score=structured_model.invoke(prompt)

    state['score']=score

    return state
    
```

```python
graph.add_node('create_outline',create_outline)
graph.add_node('create_blog',create_blog)
graph.add_node('generate_score',generate_score)
```

```python
graph.add_edge(START,'create_outline')
graph.add_edge('create_outline','create_blog')
graph.add_edge('create_blog','generate_score')
graph.add_edge('generate_score',END)
```

```python
workflow=graph.compile()
```

```python
final_state=workflow.invoke({
    'title':"Openai"
})
from pprint import pprint
pprint(final_state['score'])
```
